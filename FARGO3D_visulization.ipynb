```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '/disk2/alma/fargo3d/outputs/alma_fargo/secuencia/' # aqui tienes que poner la carpeta donde se encuentran los outputs de FARGO3D
sim = 10
path = f'/disk2/alma/pipeline/hydro/outputs/Alma_outputs/sim_{(sim):01d}/'
r = np.loadtxt(path+'domain_y.dat')[3:-4]
phi = np.loadtxt(path+'domain_x.dat')[:-1]
nr = len(r)
nphi = len(phi)

Phi, R = np.meshgrid(phi,r)
X = R*np.cos(Phi)
Y = R*np.sin(Phi)
```

```python
t = 10 # numero del output
gasdens = np.fromfile(path+'gasdens{:d}.dat'.format(t)).reshape(nr,nphi)
dustdens = np.fromfile(path+'dust1dens{:d}.dat'.format(t)).reshape(nr,nphi)
###lo mismo para las velocidades ###
```

```python
fig = plt.figure()
plt.pcolormesh(X,Y,gasdens)
ax = plt.gca()
ax.set_aspect(1)
plt.show()

fig = plt.figure()
plt.pcolormesh(X,Y,np.log(dustdens))
ax = plt.gca()
ax.set_aspect(1)
plt.show()
```
